```python
import pandas as pd
import tabula
import os
```

```python
files = os.listdir("data")
df = pd.DataFrame(columns=['Фірма', "ЄРДПОУ"])

for file in files:
    tables = tabula.read_pdf("data/" + file, pages="all")
    print("Current file: " + file)
    
    for table in tables:
        if(len(table) != 0):
            if(table.columns[2] != "Заявник"):
                next_data = table.columns.to_list()[2:4]
                next_data = pd.DataFrame([next_data], columns=df.columns)
                df = pd.concat([df, next_data], ignore_index=True)
            
            next_data = table.iloc[:, [2, 3]]
            next_data.columns = df.columns
            df = pd.concat([df, next_data], ignore_index=True)
```

```python
# Data clearing
df["ЄРДПОУ"] = pd.to_numeric(df["ЄРДПОУ"], errors="coerce")
df = df.dropna(subset="ЄРДПОУ")
```
