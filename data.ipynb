```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://minagro.gov.ua/napryamki/eksport-do-krain-ies/nakazy-z-veryfikatsii-subiektiv-ahropromyslovoho-kompleksu"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

links = soup.find_all("ul")[1].find_all('a')
# print(blocks)
for link in links:
    print(link.get('href'), end="\n")
    
```
