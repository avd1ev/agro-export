```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://minagro.gov.ua/napryamki/eksport-do-krain-ies/nakazy-z-veryfikatsii-subiektiv-ahropromyslovoho-kompleksu"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

links = soup.find_all("ul")[1].find_all('a')

for link in links:
    print(link.get('href'), end="\n")
    
```

```python
for i, link in enumerate(links):
    response = requests.get(link.get('href'))
    soup = BeautifulSoup(response.content, "html.parser")

    pdf_link = soup.find('object').get('data')
    file_path = "data/" + str(i) + ".pdf"
    
    try:
        response = requests.get(pdf_link)
        if response.status_code == 200:
            with open(file_path, 'wb') as file:
                file.write(response.content)
            print(f"Success. Path of file = {file_path}")
        else:
            print(f"Failed. Status code: {response.status_code}")
    except Exception as e:
        print(f"Error: {e}")
print("End of download!!!")
```
